```python
%cd / 

import os
from glob import glob

img_list = glob('/home/tutor/coco128/coco128/labels/train2017/*.txt')

print(len(img_list))
```

```python
two = [] #체크를 위한 리스트 생성

#여기서 val_img_list / img_list 두개를 변경하면서 내부의 다른 index를 제거해주어야 한다.
for x in img_list:
    two = [] #새로운 라벨링 데이터 .txt를 불렀을 때 초기화를 위함
    f = open(x, 'r')
    lines = f.readlines() 
   
    for line in lines: #읽어들인 모든 데이터를 한 줄씩 불러옴

        # backpack(2), bottle(13), cup(25), diningtable(26), handbag(35), keyboard(38), laptop(41), person(48) ,
        # suitcase(64), 
        
        # 해당되는 객체가 없으면 모두 제거하기 위해 check.
        if (int(line.split(' ')[0]) == 0) or (int(line.split(' ')[0]) == 24) or (int(line.split(' ')[0]) == 26) or (int(line.split(' ')[0]) == 28) or (int(line.split(' ')[0]) == 39) or (int(line.split(' ')[0]) == 41) or (int(line.split(' ')[0]) == 60) or (int(line.split(' ')[0]) == 63) or (int(line.split(' ')[0]) == 64) or (int(line.split(' ')[0]) == 65) or (int(line.split(' ')[0]) == 66) or (int(line.split(' ')[0]) == 67) or (int(line.split(' ')[0]) == 73):
            two.append(line)

    # 만약 하나도 없는 사진이 있다면 제거.
    if len(two) == 0:
        print("two 내용 없음")
        os.remove(x.split('.txt')[0].replace('labels','images')+'.jpg')
        os.remove(x)
        
    f.close()

    #두 개가 존재한다면 다시 두 개만 불러오도록 재정의.
    #다시 쓰는이유는 하나의 사진에 48 ,63 이외에 다른 값도 있는 사진을 수정하기 위함.
    #w를 이용해 재정의하면 원래 txt파일에 존재하는 모든 수치를 제거하고 처음부터 적는다. 이를 이용함.
    if len(two) >= 1:
        print("two 씀: ", len(two))
        w = open(x,'w')
        w.writelines(two)
        w.close()
```
